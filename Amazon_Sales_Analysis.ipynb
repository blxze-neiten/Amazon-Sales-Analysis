```python
# ==========================================
# AMAZON SALES ANALYSIS
# ==========================================

# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# ==========================================
# Upload Excel File
# ==========================================

from google.colab import files

uploaded = files.upload()
```

```python
# ==========================================
# Load Dataset
# ==========================================

df = pd.read_excel("amazon.xlsx")
```

```python
# ==========================================
# Answer 1
# Display the first five rows of the dataset
# ==========================================

df.head()
```

```python
# ==========================================
# Answer 2
# Display the number of rows and columns
# ==========================================

print("Number of Rows:", df.shape[0])
print("Number of Columns:", df.shape[1])
```

```python
# ==========================================
# Answer 3
# Display all column names
# ==========================================

print(df.columns.tolist())
```

```python
# ==========================================
# Answer 4
# Display dataset information
# ==========================================

df.info()
```

```python
# ==========================================
# Answer 5
# Display descriptive statistics
# ==========================================

df.describe()
```

```python
# ==========================================
# Answer 6
# Check for missing values
# ==========================================

df.isnull().sum()
```

```python
# ==========================================
# Answer 7
# Remove duplicate records
# ==========================================

df.drop_duplicates(inplace=True)

print("Duplicates removed successfully.")
```

```python
# ==========================================
# Answer 8
# Create Profit column
# ==========================================

df["Profit"] = df["TotalAmount"] * 0.20

df.head()
```

```python
# ==========================================
# Answer 9
# Calculate Total Sales
# ==========================================

total_sales = df["TotalAmount"].sum()

print("Total Sales =", total_sales)
```

```python
# ==========================================
# Answer 10
# Calculate Total Profit
# ==========================================

total_profit = df["Profit"].sum()

print("Total Profit =", total_profit)
```

```python
# ==========================================
# Answer 11
# Calculate Total Orders
# ==========================================

total_orders = df["OrderID"].nunique()

print("Total Orders =", total_orders)
```

```python
# ==========================================
# Answer 12
# Average Sales
# ==========================================

average_sales = df["TotalAmount"].mean()

print("Average Sales =", average_sales)
```

```python
# ==========================================
# Answer 13
# Average Profit
# ==========================================

average_profit = df["Profit"].mean()

print("Average Profit =", average_profit)
```

```python
# ==========================================
# Answer 14
# Maximum Sales
# ==========================================

maximum_sales = df["TotalAmount"].max()

print("Maximum Sales =", maximum_sales)
```

```python
# ==========================================
# Answer 15
# Minimum Sales
# ==========================================

minimum_sales = df["TotalAmount"].min()

print("Minimum Sales =", minimum_sales)
```

```python
# ==========================================
# Answer 16
# Total Sales by State
# ==========================================

sales_state = (
    df.groupby("State")["TotalAmount"]
      .sum()
      .sort_values(ascending=False)
)

sales_state
```

```python
# ==========================================
# Answer 17
# Total Sales by Category
# ==========================================

sales_category = (
    df.groupby("Category")["TotalAmount"]
      .sum()
      .sort_values(ascending=False)
)

sales_category
```

```python
# ==========================================
# Answer 18
# Create SubCategory
# ==========================================

def assign_subcategory(category):
    if category == "Electronics":
        return "Gadgets"
    elif category == "Clothing":
        return "Fashion"
    elif category == "Home & Kitchen":
        return "Home"
    elif category == "Books":
        return "Education"
    elif category == "Sports & Outdoors":
        return "Fitness"
    elif category == "Toys & Games":
        return "Kids"
    else:
        return "Others"

df["SubCategory"] = df["Category"].apply(assign_subcategory)

df.head()
```

```python
# ==========================================
# Answer 19
# Sales by SubCategory
# ==========================================

sales_subcategory = (
    df.groupby("SubCategory")["TotalAmount"]
      .sum()
      .sort_values(ascending=False)
)

sales_subcategory
```

```python
# ==========================================
# Answer 20
# Highest Revenue Category
# ==========================================

highest_category = sales_category.idxmax()

print(highest_category)
```

```python
# ==========================================
# Answer 21
# Highest Revenue SubCategory
# ==========================================

highest_subcategory = sales_subcategory.idxmax()

print(highest_subcategory)
```

```python
# ==========================================
# Answer 22
# Top 5 Customers
# ==========================================

top_customers = (
    df.groupby("CustomerName")["TotalAmount"]
      .sum()
      .sort_values(ascending=False)
      .head(5)
)

top_customers
```

```python
# ==========================================
# Answer 23
# Top 10 Products
# ==========================================

top_products = (
    df.groupby("ProductName")["TotalAmount"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

top_products
```

```python
# ==========================================
# Answer 24
# Orders by Payment Method
# ==========================================

payment = df["PaymentMethod"].value_counts()

payment
```

```python
# ==========================================
# Answer 25
# Monthly Sales Trend
# ==========================================

df["OrderDate"] = pd.to_datetime(df["OrderDate"])

df["Month"] = df["OrderDate"].dt.strftime("%B")

monthly_sales = df.groupby("Month")["TotalAmount"].sum()

monthly_sales
```

```python
# ==========================================
# Answer 26
# Highest Sales Month
# ==========================================

highest_sales_month = monthly_sales.idxmax()

print(highest_sales_month)
```

```python
# ==========================================
# Answer 27
# Monthly Profit
# ==========================================

monthly_profit = df.groupby("Month")["Profit"].sum()

monthly_profit
```

```python
# ==========================================
# Answer 28
# Highest Profit Month
# ==========================================

highest_profit_month = monthly_profit.idxmax()

print(highest_profit_month)
```

```python
# ==========================================
# Answer 29
# Highest and Lowest Sales State
# ==========================================

highest_state = sales_state.idxmax()

lowest_state = sales_state.idxmin()

print("Highest Sales State:", highest_state)
print("Lowest Sales State:", lowest_state)
```

```python
# ==========================================
# Answer 30
# Brand Analysis
# ==========================================

brand_sales = (
    df.groupby("Brand")["TotalAmount"]
      .sum()
      .sort_values(ascending=False)
)

brand_sales
```

```python
# ==========================================
# Answer 31
# Order Status Analysis
# ==========================================

order_status = df["OrderStatus"].value_counts()

order_status
```

```python
# ==========================================
# Answer 32
# Average Discount
# ==========================================

average_discount = df["Discount"].mean()

print(average_discount)
```

```python
# ==========================================
# Answer 33
# Average Shipping Cost
# ==========================================

average_shipping = df["ShippingCost"].mean()

print(average_shipping)
```

```python
# ==========================================
# Answer 34
# Sales by State Bar Chart
# ==========================================

sales_state.plot(
    kind="bar",
    figsize=(10,6)
)

plt.title("Sales by State")
plt.xlabel("State")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)

plt.show()
```

```python
# ==========================================
# Answer 35
# Sales by Category Pie Chart
# ==========================================

sales_category.plot(
    kind="pie",
    autopct="%1.1f%%",
    figsize=(8,8)
)

plt.title("Sales by Category")
plt.ylabel("")

plt.show()
```

```python
# ==========================================
# Answer 36
# Monthly Sales Trend
# ==========================================

monthly_sales.plot(
    kind="line",
    marker="o",
    figsize=(10,5)
)

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.grid(True)

plt.show()
```

```python
# ==========================================
# Answer 37
# Top 10 Products
# ==========================================

top_products.plot(
    kind="bar",
    figsize=(12,6)
)

plt.title("Top 10 Products")
plt.xlabel("Product")
plt.ylabel("Sales")

plt.xticks(rotation=45)

plt.show()
```

```python
# ==========================================
# Answer 38
# Profit by Category
# ==========================================

profit_category = (
    df.groupby("Category")["Profit"]
      .sum()
      .sort_values(ascending=False)
)

profit_category.plot(
    kind="bar",
    figsize=(10,5)
)

plt.title("Profit by Category")
plt.xlabel("Category")
plt.ylabel("Profit")

plt.xticks(rotation=45)

plt.show()
```

```python
# ==========================================
# Answer 39
# Orders by Payment Method
# ==========================================

payment.plot(
    kind="bar",
    figsize=(8,5)
)

plt.title("Orders by Payment Method")
plt.xlabel("Payment Method")
plt.ylabel("Number of Orders")

plt.xticks(rotation=45)

plt.show()
```

```python
# ==========================================
# Answer 40
# Top 5 Customers
# ==========================================

top_customers.plot(
    kind="bar",
    figsize=(10,5)
)

plt.title("Top 5 Customers")
plt.xlabel("Customer")
plt.ylabel("Sales")

plt.xticks(rotation=45)

plt.show()
```
